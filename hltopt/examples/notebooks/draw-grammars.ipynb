```python
class JupyterDot(pydot.Dot):
    def _repr_svg_(self):
        return self.create_svg().decode('utf8')
    
graph = JupyterDot()
```

```python
for key, prod in model.items():
    graph.add_node(pydot.Node(key, shape='rectangle'))
    
    if hasattr(prod, 'rules'):
        for rule in prod.rules:
            pass
```
