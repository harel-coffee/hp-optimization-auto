```python
from matplotlib import pyplot as plt
```

```python
import pandas as pd
import numpy as np
```

```python
import json

data = []

with open("../../../GE.json") as fp:
    for line in fp:
        data.append(json.loads(line))
```

```python
for d in data:
    d['avg'] = np.mean(d['fitness'])
    d['var'] = np.std(d['fitness'])
```

```python
df = pd.DataFrame(data)
```

```python
plt.errorbar(np.arange(0,100), df['avg'], df['var'])
plt.ylim(0,1)
```
